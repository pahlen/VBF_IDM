```python
def statist(XS_sig,XS_back,lumi,En_sig,En_back,En_sig_0,En_back_0):
    import numpy as np
    '''
    input description
    --------------------------------
    XS_sig  --> Signal cross-section [pb]
    XS_back --> Background cross-section [pb]
    lumi    --> Luminosity in [fb-1]
    En_sig  --> Entries for the signal
    En_back --> Entries for the background
    --------------------------------
    output description
    --------------------------------
    0. ws --> Signal Weight
    1. wb --> Background Weight
    2. Ns --> Signal events
    3. Nb --> Background events
    4. R1 --> Ns/Nb ratio
    5. Sig --> Significance
    6. Eff_sig --> Signal efficiency 
    7. Eff_backg --> background efficiency 
    '''
    ws=(lumi*XS_sig*1000.)/En_sig_0
    wb=(lumi*XS_back*1000.)/En_back_0
    Ns=En_sig*ws
    Nb=En_back*wb
    R1=Ns/Nb
    Sig=Ns/np.sqrt(Ns+Nb)
    Eff_sig = En_sig/En_sig_0
    Eff_backg = En_back/En_back_0
    return ws,wb,Ns,Nb,R1,Sig,Eff_sig,Eff_backg
```
